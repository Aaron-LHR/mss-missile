```python
def data_preprocess(single_trace, mu, sigma):
    point_list = []
    # value_list = []
    start_point_gps = single_trace[-1][:3]
    end_point_gps = single_trace[-1][4:7]
    # value_list.extend(start_point_gps)
    # value_list.extend(end_point_gps)
    for row in single_trace[:-2]:
        cur_xyz = row[1:4]
        # value_list.extend(cur_xyz)
        point_list.append(np.stack([start_point_gps, cur_xyz, end_point_gps]))
    points_numpy = np.stack(point_list)
    # mu = np.mean(value_list)
    # sigma = np.std(value_list)
    points_numpy = (points_numpy - mu) / sigma
    return points_numpy[:-1], points_numpy[1:, 1:2].squeeze()
```

```python
class BaseLine(torch.nn.Module):
    def __init__(self, input_size=240, hidden_size=320, output_size=3):
        super().__init__()
        self.batch_first = True
        self.cnn1 = torch.nn.Conv2d(1, 32, kernel_size=1)
        self.cnn2 = torch.nn.Conv2d(32, 32, kernel_size=1)
        self.cnn3 = torch.nn.Conv2d(32, 80, kernel_size=1)
        self.cnn4 = torch.nn.Conv2d(80, 80, kernel_size=1)

        self.gru = torch.nn.GRU(input_size=240, hidden_size=320)
        self.lstm = torch.nn.LSTM(input_size, hidden_size, 2,
                                  batch_first=self.batch_first)  # (input_size, hidden_size, num_layers)

        self.linear1 = nn.Linear(hidden_size, 160)
        self.linear2 = nn.Linear(160, 320)
        self.linear3 = nn.Linear(320, 288)
        self.linear4 = nn.Linear(288, 3)

    def forward(self, input_tensor, seq_lens=[]):
        x = input_tensor
        # print(x.shape)
        x = self.cnn1(x)
        # print(x.shape)
        x = torch.nn.functional.relu(x)
        # print(x.shape)
        x = self.cnn2(x)
        # print(x.shape)
        x = torch.nn.functional.relu(x)
        # print(x.shape)
        x = torch.nn.functional.max_pool2d(x, kernel_size=(3, 1), padding=(0, 0))

        # print(x.shape)
        x = self.cnn3(x)
        # print(x.shape)
        x = torch.nn.functional.relu(x)
        # print(x.shape)
        x = self.cnn4(x)
        # print(x.shape)
        x = torch.nn.functional.relu(x)
        # print(x.shape)
        x = torch.nn.functional.max_pool2d(x, kernel_size=(1, 1), padding=(0, 0))

        # print(x.shape)
        x = x.view(-1, 240)
        # print(x.shape)
        x = x.unsqueeze(1)

        # input_tensor shape: batch_size * time_step * dim , seq_lens: (batch_size,)  when batch_first = True
        # total_length = input_tensor.size(1) if self.batch_first else input_tensor.size(0)
        # x_packed = pack_padded_sequence(input_tensor, seq_lens, batch_first=self.batch_first, enforce_sorted=False)
        # y_lstm, hidden = self.lstm(x)
        # y_padded, length = pad_packed_sequence(y_lstm, batch_first=self.batch_first, total_length=total_length)

        # print(x.shape)
        out, hidden = self.gru(x)

        # print(x.shape)
        x = self.linear1(out)
        # print(x.shape)
        x = self.linear2(x)
        # print(x.shape)
        x = self.linear3(x)
        # print(x.shape)
        x = self.linear4(x)

        return x
```

```python
def get_mu_and_sigma(trace_list):
    value_list = []
    for single_trace in trace_list:
        start_point_gps = single_trace[-1][:3]
        end_point_gps = single_trace[-1][4:7]
        value_list.extend(start_point_gps)
        value_list.extend(end_point_gps)
        for row in single_trace[:-2]:
            cur_xyz = row[1:4]
            value_list.extend(cur_xyz)
    mu = np.mean(value_list)
    sigma = np.std(value_list)
    return mu, sigma
```

```python
class Curve_Model_2d(torch.nn.Module):
    def __init__(self):
        super().__init__()
        self.linear1 = nn.Linear(1, 128)
        self.linear2 = nn.Linear(128, 256)
        self.linear3 = nn.Linear(256, 1)

    def forward(self, x):
        x = self.linear1(x)
        x = torch.nn.functional.relu(x)
        x = self.linear2(x)
        x = torch.nn.functional.relu(x)
        x = self.linear3(x)
        x = torch.nn.functional.relu(x)

        return x
```

```python
def get_geocentric_coordinate_system(Lt, Bt, Ht):  # 经纬高
    a = 6378137  # 地球椭球长半轴
    e = 0.003352813177897

    Lt = Lt * np.pi / 180
    Bt = Bt * np.pi / 180

    Nt = a * (1 + e * np.sin(Bt) * np.sin(Bt))  # 当地卯酉圈曲率半径

    Xt = (Nt + Ht) * np.cos(Bt) * np.cos(Lt)  # 地心坐标
    Yt = (Nt + Ht) * np.cos(Bt) * np.sin(Lt)
    Zt = (Nt * (1 - e) * (1 - e) + Ht) * np.sin(Bt)

    return Xt, Yt, Zt
```

```python
def visualize_trace(trace, figure_index, x_index=7):
    fig = plt.figure(figsize=(8, 4))
    ax = fig.add_subplot(projection='3d')
    # 轨迹
    for i in range(len(trace[x_index][:-2])):
        trace[x_index][:-2][i], trace[x_index + 1][:-2][i], trace[x_index + 2][:-2][i] = get_geocentric_coordinate_system(trace[x_index][:-2][i], trace[x_index + 1][:-2][i], trace[x_index + 2][:-2][i])
    ax.plot(xs=trace[x_index][:-2], ys=trace[x_index + 1][:-2], zs=trace[x_index + 2][:-2])
    # 起点
    ax.scatter(trace[x_index][:1], trace[x_index + 1][:1], trace[x_index + 2][:1], 'r')
    # 飞行姿态变化
    first_change_time = trace[0][-2]
    first_change_time_index = int(first_change_time // 0.05)
    second_change_time = trace[1][-2]
    second_change_time_index = int(second_change_time // 0.05)
    # print(trace.shape)
    # print(first_change_time)
    # print(first_change_time_index)
    # print(second_change_time)
    # print(second_change_time_index)
    ax.scatter(trace[x_index][first_change_time_index], trace[x_index + 1][first_change_time_index],
               trace[x_index + 2][first_change_time_index], 'r')
    ax.scatter(trace[x_index][second_change_time_index], trace[x_index + 1][second_change_time_index],
               trace[x_index + 2][second_change_time_index], 'r')
    fig.savefig(f"./figure/geocentric/trace_{figure_index}.png", dpi=300)
    fig.show()
```

```python
def pedal(p1, p2, p3):
    """
    过p3作p1和p2相连直线的垂线, 计算垂足的坐标
    直线1：垂足坐标和p3连线
    直线2: p1和p2连线
    两条直线垂直, 且交点为垂足
    :param p1: (x1, y1)
    :param p2: (x2, y2)
    :param p3: (x3, y3)
    :return: 垂足坐标 (x, y)
    """
    if p2[0] != p1[0]:
        # ########## 根据点x1和x2计算线性方程的k, b
        k, b = np.linalg.solve([[p1[0], 1], [p2[0], 1]], [p1[1], p2[1]])  # 得到k和b
        # #######原理: 垂直向量数量积为0
        x = np.divide(((p2[0] - p1[0]) * p3[0] + (p2[1] - p1[1]) * p3[1] - b * (p2[1] - p1[1])),
                      (p2[0] - p1[0] + k * (p2[1] - p1[1])))
        y = k * x + b

    else:  # 点p1和p2的连线垂直于x轴时
        x = p1[0]
        y = p3[1]

    return x, y


if __name__ == '__main__':
    x0, y0 = pedal((20, 60), (120, 100), (70, 30))
    print(f"垂足的坐标是 {x0, y0}")

```

```python
def get_y_side_symbol(a, b, t):
    return np.sign((a[0]-t[0])*(b[1]-t[1])-(a[1]-t[1])*(b[0]-t[0]))
```

```python
def get_x_y_z(single_line_data, x_list, yz_list, x_index=7):
    start_point = single_line_data[x_index: x_index + 3, 0]
    end_point = single_line_data[x_index: x_index + 3, -3]

    span = np.linalg.norm(end_point - start_point)
    for i in range(single_line_data.shape[1] - 2):
        cur_point = single_line_data[x_index: x_index + 2, i]
        pedal_point_xy = pedal(start_point, end_point, cur_point)
        pedal_point = np.array([pedal_point_xy[0], pedal_point_xy[1], 0])
        xy_project_point = np.array([cur_point[0], cur_point[1], 0])
        x = np.linalg.norm(pedal_point - start_point)
        y = np.linalg.norm(xy_project_point - pedal_point) * get_y_side_symbol(start_point, end_point, xy_project_point)
        z = single_line_data[x_index + 2, i]

        x_list.append(np.array([x.item(), span.item()], dtype=np.float32))
        yz_list.append(np.array([y.item(), z.item()], dtype=np.float32))

    return x_list, yz_list
```

```python
# %%
import h5py
import numpy as np
import matplotlib.pyplot as plt
from sklearn.cluster import KMeans

x_index = 7


def load_data():
    mat = h5py.File('./data/qian-data.mat')
    data = [np.transpose(mat[h5py.h5r.get_name(mat['BB'][i][j], mat.id)]) for i in range(len(mat['BB'])) for j in
            range(len(mat['BB'][0]))]
    return data


def visualize_trace(data):
    for i, line in enumerate(data):
        fig = plt.figure()
        ax = fig.add_subplot(projection='3d')
        ax.plot(xs=line[x_index], ys=line[x_index + 1], zs=line[x_index + 2])
        ax.scatter(line[x_index][:1], line[x_index + 1][:1], line[x_index + 2][:1], 'r')
        fig.show()
        # fig.savefig(f"./figure/trace_{i}.png", dpi=300)


def visualize_point(data):
    for i, line in enumerate(data):
        fig = plt.figure()
        plt.scatter(line[x_index], line[x_index + 1])
        fig.show()
        # fig.savefig(f"./figure/trace_{i}.png", dpi=300)


def visualize_2D_trace(data):
    for i, line in enumerate(data):
        fig = plt.figure()
        plt.plot(get_xy_z(line, line[:, :1]), line[x_index + 2])
        fig.show()
        # fig.savefig(f"./figure/trace_{i}.png", dpi=300)


def get_xy_z(single_line_data, start_point):
    return ((single_line_data[x_index] - start_point[x_index][0]) ** 2 + (
            single_line_data[x_index + 1] - start_point[x_index + 1][0]) ** 2) ** 0.5


def get_curve_function(line_data, deg, start_point):
    param = np.polyfit(get_xy_z(line_data, start_point), line_data[x_index + 2], deg)
    # print(param)
    curve_function = np.poly1d(param)
    # print(curve_function)
    return param, curve_function


def compare_fit_curve(real_curve, fitted_curve_list):
    fig = plt.figure()
    color_list = ['r', 'g', 'gray']
    plt.plot(real_curve[0], real_curve[1], 'b')
    for i, fitted_curve in enumerate(fitted_curve_list):
        x, y = fitted_curve
        plt.plot(list(x), list(y), color_list[i % len(color_list)])
    fig.show()


def cluster_three(data_list):
    kmeans = KMeans(
        n_clusters=n_cluster,
        n_init=10,
        max_iter=3000,
        algorithm='auto').fit(data_list)
    return kmeans.cluster_centers_


# %%
if __name__ == '__main__':
    # %%
    trace_data = load_data()
    # %%
    visualize_trace(trace_data)
    # %%
    visualize_point(trace_data)
    # %%
    visualize_2D_trace(trace_data)
    # %%
    get_curve_function(trace_data[0], 3, trace_data[0][:, :1])
    # %%
    n_cluster = 3
    fit_step_length = 300
    allowed_diff = 1000
    deg = 3
    for i, single_data in enumerate(trace_data[:1]):
        param_list = []
        for j in range(single_data.shape[1] - (fit_step_length - 1)):
            example_data = single_data[:, j: j + fit_step_length]
            param, fit_function_ = get_curve_function(example_data, deg, single_data[:, :1])
            param_list.append(param)

        fit_function_list = [np.poly1d(param) for param in param_list]
        cluster_param_center = cluster_three(param_list)
        cluster_fit_function_list = [np.poly1d(param) for param in cluster_param_center]
        xy = get_xy_z(single_data, single_data[:, :1])

        fit_curve_list = []
        for fit_function in cluster_fit_function_list:
            curve_tuple = ([], [])
            enter = False
            for p_index, p_xy in enumerate(xy):
                fit_z = fit_function(p_xy)
                if abs(fit_z - single_data[x_index + 2][p_index]) < allowed_diff:
                    enter = True
                    curve_tuple[0].append(p_xy)
                    curve_tuple[1].append(fit_z)
                elif enter:
                    break
            fit_curve_list.append(curve_tuple)
        compare_fit_curve((xy, single_data[x_index + 2]), fit_curve_list)
        # print([zip(*[(point_xy, np.poly1d(param)(point_xy)) for point_index, point_xy in enumerate(xy) if abs(np.poly1d(param)(point_xy) - single_data[x_index + 2][point_xy]) < 10]) for param in cluster_param_center])

        # compare_fit_curve((xy, single_data[x_index + 2]), [zip(*[(point_xy, fit_function(point_xy)) for point_index, point_xy in enumerate(xy) if abs(fit_function(point_xy) - single_data[x_index + 2][point_index]) < allowed_diff]) for fit_function in fit_function_list[::1000]])

# %%


```
